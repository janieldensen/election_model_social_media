# Visualizations

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
avg_sent = pd.read_csv('datasets/processed_data.csv')
avg_sent.rename({'Unnamed: 0': 'candidates'}, axis=1, inplace=True)
avg_sent.set_index('candidates', drop=True, inplace=True)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(data=avg_sent[avg_sent.party == 'd'], x='weighted_mean', y='result', color='navy', label='democrat')
plt.scatter(data=avg_sent[avg_sent.party == 'r'], x='weighted_mean', y='result', color='maroon', label='republican')
plt.axline((0,0), (100, 100), color='black')
plt.title('Scatterplot of Weighted Means and Election Results')
plt.xlabel('Weighted Mean')
plt.ylabel('Election Result')
plt.legend()
plt.xlim(30, 70)
plt.ylim(30, 70)
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')
plt.savefig('./images/weighted_means_scatter.png', dpi=192, facecolor='white');
```

```python
plt.figure(figsize=(10,10))
plt.scatter(data=avg_sent[avg_sent.party == 'd'], x='sentiment', y='result', label='democrats', color='navy')
plt.scatter(data=avg_sent[avg_sent.party == 'r'], x='sentiment', y='result', label='republicans', color='maroon')
plt.xlabel('Average Sentiment')
plt.ylabel('Election Result')
plt.title('Sentiment v Results')
plt.legend()
plt.savefig('./images/sentiment_scatter.png', dpi=192, facecolor='white');
```

```python
plt.figure(figsize=(10,8))
plt.bar(avg_sent.sort_values('sentiment').head().index, height=avg_sent.sort_values('sentiment').head().sentiment, color=['maroon','maroon','maroon','maroon','navy'])
plt.xlabel('Candidates')
plt.ylabel('Sentiment')
plt.title('Bottom 5 Candidates')
plt.savefig('./images/bottom5.png', dpi=192, facecolor='white');
```

```python
plt.figure(figsize=(10,8))
plt.bar(avg_sent[avg_sent.sentiment < 0.4].sort_values('sentiment', ascending=False).head().index, 
        height=avg_sent[avg_sent.sentiment < 0.4].sort_values('sentiment', ascending=False).head().sentiment, 
        color=['maroon','navy','maroon','navy','navy'])
plt.xlabel('Candidates')
plt.ylabel('Sentiment')
plt.title('Top 5 Candidates')
plt.savefig('./images/top5.png', dpi=192, facecolor='white');
```
